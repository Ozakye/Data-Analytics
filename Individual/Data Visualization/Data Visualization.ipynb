```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from statsmodels.tsa.seasonal import seasonal_decompose
```

#1

Carregar os dados de um arquivo CSV fornecido (saude.csv)

```python
df_saude = pd.read_csv('saude.csv')
df_saude.head()
```

#2

Criar um histograma da variável 'age' usando Matplotlib

```python
plt.style.use('bmh')
plt.hist(df_saude['age'], color='skyblue', edgecolor='black')
plt.title('Histograma da Idade')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.grid(True)
plt.show()
```

#3

Criar um gráfico de dispersão mostrando a relação entre 'height' e 'weight' usando Seaborn.

```python
plt.style.use('bmh')
sns.scatterplot(data=df_saude, x='height', y='weight')
plt.title('Relação entre Altura e Peso')
plt.xlabel('Altura')
plt.ylabel('Peso')
plt.grid(True)
plt.show()
```

#4

Criar um gráfico interativo de barras mostrando a contagem de indivíduos por
'gender' usando Plotly

```python
fig = px.bar(df_saude, x='gender', title='Contagem de Indivíduos por Gênero')
fig.show()
```

#5

Carregar os dados de um arquivo CSV fornecido (vendas.csv)

```python
df_vendas = pd.read_csv('vendas.csv')
df_vendas.head()
```

#6

Criar um gráfico de linhas das vendas ao longo do tempo usando Matplotlib

```python
plt.plot(df_vendas['date'], df_vendas['sales'], marker='o', linestyle='-')
plt.title('Vendas ao Longo do Tempo')
plt.xlabel('Data')
plt.ylabel('Vendas')
plt.grid(True)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

#7

Criar um gráfico de decomposição sazonal usando Seaborn

```python
sns.set_style('darkgrid')
result = seasonal_decompose(df_vendas['sales'], model='additive', period=12)
fig, axs = plt.subplots(4, 1, figsize=(12, 9), sharex=True)

axs[0].plot(result.observed)
axs[0].set_title('Observado')

axs[1].plot(result.trend)
axs[1].set_title('Tendência')

axs[2].plot(result.seasonal)
axs[2].set_title('Sazonalidade')

axs[3].plot(result.resid)
axs[3].set_title('Resíduo')

plt.tight_layout()
plt.show
```

#8

Criar um gráfico interativo que permita a exploração dos dados de vendas ao longo do tempo usando Plotly

```python
fig = px.line(df_vendas, x='date', y='sales', title='Vendas ao Longo do Tempo')
fig.show()
```
