#1

Crie uma tabela chamada "alunos" com os seguintes campos: id
(inteiro), nome (texto), idade (inteiro) e curso (texto).

```python
import sqlite3

con = sqlite3.connect('alunos')
cur = con.cursor()

cur.execute ('''
CREATE TABLE IF NOT EXISTS alunos (
  id INTEGER PRIMARY KEY AUTOINCREMENT,
  nome TEXT NOT NULL,
  idade INT NOT NULL,
  curso TEXT NOT NULL
)''')

con.commit()
con.close()
```

#5

Crie uma tabela chamada "clientes" com os campos: id (chave primária), nome (texto), idade (inteiro) e saldo (float). Insira alguns registros de clientes na tabela.

```python
con = sqlite3.connect('clientes')
cur = con.cursor()

cur.execute ('''
CREATE TABLE IF NOT EXISTS clientes (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nome TEXT NOT NULL,
    idade INTEGER NOT NULL,
    saldo REAL NOT NULL
);
''')

con.commit()
con.close()
```

```python
con = sqlite3.connect('clientes')
cur = con.cursor()

cur.executemany('''
INSERT INTO clientes (nome, idade, saldo)
VALUES (?, ?, ?)
''', [
    ('João Silva', 30, 1500.75),
    ('Maria Martins', 25, 2000.50),
    ('Carlos Souza', 28, 750.30),
    ('Ana Paula', 35, 3200.60)
])

cur.execute("SELECT * FROM clientes")
dados = cur.fetchall()

for cliente in dados:
  print(cliente)

con.commit()
con.close()
```

#6

Escreva consultas SQL para realizar as seguintes tarefas:

*   Selecione o nome e a idade dos clientes com idade superior a
30 anos.
*   Calcule o saldo médio dos clientes.
*   Encontre o cliente com o saldo máximo.
*   Conte quantos clientes têm saldo acima de 1000.

```python
con = sqlite3.connect('clientes')
cur = con.cursor()

cur.execute("SELECT nome, idade FROM clientes WHERE idade > 30")
dados = cur.fetchall()

for cliente in dados:
  print(cliente)

con.commit()
con.close()
```

```python
con = sqlite3.connect('clientes')
cur = con.cursor()

cur.execute("SELECT AVG(saldo) AS saldo_medio FROM clientes")
dados = cur.fetchall()

for cliente in dados:
  print(f'{cliente[0]:.2f}')

con.commit()
con.close()
```

```python
con = sqlite3.connect('clientes')
cur = con.cursor()

cur.execute("SELECT nome, saldo FROM clientes WHERE saldo = (SELECT MAX(saldo) FROM clientes)")
dados = cur.fetchall()

for cliente in dados:
  print(cliente)

con.commit()
con.close()
```

```python
con = sqlite3.connect('clientes')
cur = con.cursor()

cur.execute("SELECT COUNT(*) AS clientes_saldo_acima_1000 FROM clientes WHERE saldo > 1000")
dados = cur.fetchall()

for cliente in dados:
  print(cliente)

con.commit()
con.close()
```
