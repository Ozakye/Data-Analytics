##PlantGrowth

```python
df_plantgrowth = data('PlantGrowth')
df_plantgrowth.head()
```

1. Calcule a média do peso (weight) para cada grupo (group)

```python
group_weight_mean = df_plantgrowth.groupby('group')['weight'].mean().apply(lambda x: f'{x:.2f}')
group_weight_mean
```

2. Verifique se algum grupo tem peso médio maior que 6

```python
for group, weight in group_weight_mean.items():
    if float(weight) > 6:
        print(f"O grupo '{group}' tem um peso médio maior que 6.")
    else:
        print(f"O grupo '{group}' não tem um peso médio maior que 6.")
```

3. Crie uma tabela mostrando o peso máximo e mínimo por grupo

```python
df_plantgrowth.groupby('group')['weight'].agg(['max', 'min']).applymap(lambda x: f'{x:.2f}')
```

##HairEyeColor

```python
df_hairEyeColor = data('HairEyeColor')
df_hairEyeColor
```

1. Calcule o número total de pessoas para cada combinação de cor de
cabelo (Hair) e cor dos olhos (Eye)

```python
# Filtrar os dados de frequencia da combinação de cor e cabelo
df_HairEyeColor = df_hairEyeColor.groupby(['Hair', 'Eye'])['Freq'].sum().reset_index()
df_HairEyeColor
```

2. Encontre a cor de cabelo mais comum entre pessoas com olhos
castanhos (Brown)

```python
# Agrupar por cor de cabelo e cor Olho Marrom
df_brown_eyes = df_hairEyeColor[df_hairEyeColor['Eye'] == 'Brown']
#print(df_brown_eyes)

# Soma as frequencias acima e mostra o index que contém o maior valor
hair = df_brown_eyes.groupby('Hair')['Freq'].sum().idxmax()
hair
```

3. Crie uma tabela com a contagem total de pessoas por cor de cabelo

```python
#Tabela com a frequencia de todos os tipos de cabelo
hair_all = df_hairEyeColor.groupby('Hair')['Freq'].sum().reset_index()
hair_all
```

#5
##Chickwts

```python
df_chickwts = data('chickwts')
df_chickwts.head()
```

```python
#Traduzindo Colunas
df_chickwts.rename(columns={
    'weight': 'Peso',
    'feed': 'Alimentação'
}, inplace=True)
df_chickwts.head()
```

```python
# Traduzindo itens
dicionário = {
    'horsebean': 'Feijão-cavalo',
    'linseed': 'Linhaça',
    'soybean': 'Soja',
    'sunflower': 'Girassol',
    'meatmeal': 'Farinha de carne',
    'casein': 'Caseína'
}

df_chickwts['Alimentação'] = df_chickwts['Alimentação'].replace(dicionário)

df_chickwts.head()
```

1. Calcule o peso médio dos pintinhos (weight) para cada tipo de
ração (feed)

```python
média_pesos = df_chickwts.groupby('Alimentação')['Peso'].mean().reset_index()

média_pesos['Peso'] = média_pesos['Peso'].map('{:.2f}'.format)

print('A média dos pesos dos pintinhos para cada tipo de ração é:')
média_pesos
```

2. Identifique a ração que resultou no maior peso médio

```python
ração_maior_peso = média_pesos.loc[média_pesos['Peso'].astype(float).idxmax()]
print(f"A ração que resultou em maior peso foi: {ração_maior_peso['Alimentação']} com peso médio de {ração_maior_peso['Peso']} gramas.")
```

3. Crie uma tabela que mostre o peso máximo, mínimo e médio
para cada tipo de ração

```python
tabela_calculo = df_chickwts.groupby('Alimentação')['Peso'].agg(Máximo='max', Mínimo='min', Média='mean').reset_index()
print(tabela_calculo)
```

##Attitude

```python
df_attitude = data('attitude')
df_attitude.head()
```

1. Calcule a média de "aprendizado" (learning)
por nível de
motivação (rating)

```python
#Agrupando por 'rating' e calculando a média de 'learning'
media_learning_por_rating = df_attitude.groupby('rating')['learning'].mean().reset_index()
media_learning_por_rating
```

2. Encontre a combinação de rating e complaints que possui o
maior valor médio de aprendizado

```python
#Agrupando por 'rating' e 'complaints' e calculando média de 'learning'
combinacao_rating_complaints = df_attitude.groupby(['rating', 'complaints'])['learning'].mean().reset_index()
combinacao_rating_complaints_max = combinacao_rating_complaints.loc[combinacao_rating_complaints['learning'].idxmax()]

combinacao_rating_complaints_max
```

3. Qual é o nível médio de aprendizado para funcionários que
receberam aumentos (raises) acima da média?

```python
#Calculando média de 'raises'
media_raises = df_attitude['raises'].mean()

#Filtrando funcionários com 'raises' acima da média
acima_media = df_attitude[df_attitude['raises'] > media_raises]

#Calculando média de 'learning' desses funcionários
media_learning_acima_raises = acima_media['learning'].mean()

media_learning_acima_raises
```

##Titanic

```python
df_titanic = data('titanic')
df_titanic.head()
```

1. Qual a taxa de sobrevivência geral no navio?

```python
df_titanic["survived"] = df_titanic["survived"].replace({"yes": 1, "no": 0}).astype(int)
taxa_sobrevivencia_geral = df_titanic["survived"].mean() * 100
taxa_sobrevivencia_geral_formatada = f"{taxa_sobrevivencia_geral:.2f}"
taxa_sobrevivencia_geral_formatada
```

2. Qual a taxa de sobrevivência por classe (Pclass)?

```python
taxa_sobrevivencia_classe = df_titanic.groupby("class")["survived"].mean() * 100
taxa_sobrevivencia_classe_formatada = taxa_sobrevivencia_classe.apply(lambda x: f"{x:.2f}")
taxa_sobrevivencia_classe_formatada
```

3. Qual o sexo com maior taxa de sobrevivência?

```python
taxa_sobrevivencia_sexo = df_titanic.groupby("sex")["survived"].mean() * 100
sexo_maior_sobrevivencia = taxa_sobrevivencia_sexo.idxmax()
sexo_maior_sobrevivencia
```

4. Baseado nos dados informados. Qual a quantidade de adultos e
crianças (sobreviventes) e não sobreviventes?

```python
print(df_titanic["age"].unique())
resultados = df_titanic.groupby(["age", "survived"]).size().unstack(fill_value=0)
resultados.columns = ["Não Sobreviventes", "Sobreviventes"]
print(resultados)
```
